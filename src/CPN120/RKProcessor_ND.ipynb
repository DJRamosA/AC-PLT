```python
import pandas as pd
import numpy as np
import spacy
```

```python
# Importar datos
data = pd.read_csv(r'../../data/nueva_data/propiedades_codigo.csv', delimiter="\t")
```

```python
data_matrix = pd.concat(
    (data.iloc[:,0],    # Cue/Concept
    data.iloc[:,2],     # Codification
    data.iloc[:,1]      # Description of the cue
    ),
    axis = 1
)
data_matrix
```

```python
# Transformacion de datos necesaria para aplicar el algoritmo de RK-Processor
concept_cluster = {i:data_matrix.iloc[:,2][data_matrix.iloc[:,0]==i] for i in np.unique(data_matrix.iloc[:,0])}

maximo = 0
for key,value in concept_cluster.items():
    temp = len(value)
    if temp>maximo: maximo = temp
print(maximo)

for key,value in concept_cluster.items():
    temp = len(value)
    if temp<maximo:
        concept_cluster[key] = np.append(value,['']*(maximo-temp),axis=0)
```

```python
transform_data = pd.DataFrame(concept_cluster)
transform_data.to_csv(r'../../data/nueva_data/RK_NewData.csv')
```

# Datos procesados

```python
data_processed = pd.read_csv(r'../data/RK_processed_data.csv')
```

```python
import spacy

nlp = spacy.load('es_core_news_sm')
for i in range(data_processed.shape[0]):
    for j in range(data_processed.shape[1]):
        if type(data_processed.iloc[i,j]) == str :
            doc = nlp(data_processed.iloc[i,j])

            words = [t.orth_ for t in doc if not t.is_punct]
            data_processed.iloc[i,j] = " ".join(words)
```

```python
codification = {i:data_matrix.iloc[:,1][data_matrix.iloc[:,0]==i] for i in np.unique(data_matrix.iloc[:,0])}

for key,value in codification.items():
    temp = len(value)
    if temp<maximo:
        codification[key] = np.append(value,['']*(maximo-temp),axis=0)
```

```python
original_code_data = pd.DataFrame(codification)
original_code_data
```

```python
# Generation of test-train data

from sklearn.model_selection import KFold

kfold=KFold(n_splits=5)

len_test = []
test_codes = {}
i=1
for train_index, test_index in kfold.split(data_matrix):
    test = data_matrix.iloc[test_index,:]
    len_test.append(len(test_index))
    # Transformacion de datos necesaria para aplicar el algoritmo de RK-Processor
    concept_cluster_test = {i:test.iloc[:,2][test.iloc[:,0]==i] for i in np.unique(test.iloc[:,0])}

    maximo = 0
    for key,value in concept_cluster_test.items():
        temp = len(value)
        if temp>maximo: maximo = temp

    for key,value in concept_cluster_test.items():
        temp = len(value)
        if temp<maximo:
            concept_cluster_test[key] = np.append(value,['']*(maximo-temp),axis=0)

    codification_test = {i:test.iloc[:,1][test.iloc[:,0]==i] for i in np.unique(test.iloc[:,0])}

    for key,value in codification_test.items():
        temp = len(value)
        if temp<maximo:
            codification_test[key] = np.append(value,['']*(maximo-temp),axis=0)

    transform_data = pd.DataFrame(concept_cluster_test)
    transform_data.to_csv(r'../../data/nueva_data/rk_data/RK_data_test{}.csv'.format(i))

    test_codes[i]=pd.DataFrame(codification_test)
    i+=1
```
